## 1 A simple plot

```python
from matplotlib import pyplot as plt
```

### 1.1 Letâ€™s look at some code.

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.2 Adding another plot

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.3 Yet another plot but with error bars

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.errorbar(x, y_3, xerr = x_error, yerr = y_error, 
             color = "green", label = "Y3 with errors")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.6 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend()
plt.show()
```

### 1.7 We need another axis!

```python
x = np.array([0,1,2,3,4,5,6])
x=np.linspace(0,6,100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x,x2,color='red', ls='dashed', label='x^2')  #Color not colour
plt.legend(loc='lower left')

plt.twinx() #creates another y-axis for the plot
plt.plot(x, sin_x, c='blue', label='sin(x)')
plt.legend(loc='lower right')
plt.show()
```

### 1.8 Saving to disc

```python
x = np.array([0,1,2,3,4,5,6])
x=np.linspace(0,6,100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x,x2,color='red', ls='dashed', label='x^2')  #Color not colour
plt.legend(loc='lower left')

plt.twinx() #creates another y-axis for the plot
plt.plot(x, sin_x, c='blue', label='sin(x)')
plt.legend(loc='lower right')
plt.savefig('simple-plot.png', dpi=150)
```

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
# plt.style.use('default') Reset the style
plt.style.use('fivethirtyeight')
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:,0]
anomaly = data[:,-1]

plt.plot(date, anomaly, alpha=0.5)
plt.hlines(0, date[0], date[-1], ls='dashed', color='grey')
plt.ylim([-8,8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to average from Jan 1951 - Dec 1980.)')
plt.show()
```

## Exercise 2 :  A simple mathematical plot

```python
x = np.linspace(0, 50, 1000)
y = np.exp(-x/10)*np.sin(x)
plt.plot(x, y, label='y value')
plt.ylim(-1,1)
plt.xlabel('x', fontsize=15, color='darkblue')
plt.ylabel('y', rotation=0, fontsize=15, color='darkblue')
plt.title(r'$y$ = $e^{-x/10}sin(x)$', fontsize=20)
plt.grid(alpha=0.25)
plt.show()
```
