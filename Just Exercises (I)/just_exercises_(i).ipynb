## Exercise 4 :  A function for Fibonacci

```python
def fibonacci(n):
    fibonnaci_list=[0, 1]
    for i in range(n-2):
        sum_last_two = sum(fibonnaci_list[-2:])
        fibonnaci_list.append(sum_last_two)       
    if n == 2:
        return fibonnaci_list[0:2]
    elif n==1:
        return fibonnaci_list[0]
    elif n<0:  
        return 'The number is negative.'
    else:
        return fibonnaci_list
fibonacci(5)
```

## Exercise 5 :  max_info() without NumPy

```python
numbers = [40, 27, 83, 44, 74, 51, 76, 77, 10, 49]
def max_info(numbers):
    my_max = numbers[0]
    my_location = 0
    for i in numbers:
        if i < my_max: continue
        my_max = i 
    for y in numbers:
        if y == my_max: break
        my_location += 1            
    return my_max,my_location
x,y = max_info(numbers)
        

```
